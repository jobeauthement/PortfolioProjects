```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'C:\Users\jobea\Downloads\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# let's see if there is any missing data


for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing * 100))
```

```python
# Drop the missing data

df = df.dropna(how = 'any', axis = 0)
```

```python
# Check to see that there is no missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Changing the columns data type from float to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct Year column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', 20)
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
